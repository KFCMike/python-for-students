


## Matplotlib
Today we will go through graph plotting in Python. In this tutorial, we will be using a famous libary _Matplotlib_ to demostrate how graph plotting is done in jupyter. _Matplotlib_ is a common plotting library which generates 2D graphs and works well in both jupyter and IPython. The graph is in good quality which can be included into technical report and thesis. _Matplotlib_ is designed to be simple and you can plot one graph with one single command.

## To import Matplotlib
We usually import `matplotlib.pyplot` which is a collection of related plotting function. The function design is based on Matlab. Each function changes the properties of a figure, for example, data, look and feel and title etc.

```python
import matplotlib.pyplot as plt
```

# Line Chat
Now let's try to plot a simple line chart first. There are only three commands below.

- 1st line to put data into the plot,
- 2nd line to update the label in y-axis and
- 3rd line to show the graph in jupyter.

Isn't that simple :p ?

```python
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

Now, let's try another line chart. In the below graph, we have different values in both x-axis and y-axis. Also, we can set a title to a graph!

```python
plt.plot([1, 2, 3, 4], [0, 1, 0, 1])
plt.ylabel('1 - 4')
plt.xlabel('0, 1')
plt.title('2nd Line Chart')
plt.show()
```

# Bar Chart
We just plot the first kind of graphs in jupyter, now let's try bar chart!

```python
y = [1, 2, 3, 4, 4]
x = [-1.0, 1.0, 2.0, 1.5, 3.0]
plt.bar(x, y)
plt.show()
```

# Pie Chart

```python
numbers = [1, 2, 3, 4 ,5]
labels = ["A", "B", "C", "D", "E"]
plt.title('Sample Pie Chart')
plt.pie(numbers, labels=labels, autopct='%1.1f%%', shadow=True)
plt.show()
```

# Scatter

```python
import random
x = range(0, 20)
y = [random.uniform(0, 10) for a in x]
x2 = [3.14 * a * a * 10 for a in x]
area =  x2
colors = x
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

# Sankey

```python
from matplotlib.sankey import Sankey
flows = [0.5, 1.1, -0.3, 0.4, -2.0]
labels = ['Source A', 'Source B', 'Source C', 'Source D', 'Source E']
orientations = [1, 1, -1 , 1, 0]
Sankey(flows=flows,
       labels=labels,
       orientations=orientations).finish()
plt.show()
```

# Graph Customization

## Subplot

```python
plt.plot([1,2,3])
plt.subplot(211)
plt.scatter([1, 2, 3], [4 ,6 , 2])
plt.subplot(212, facecolor='y')
plt.show()
```

## No border

```python
y = [1, 2, 3, 4, 4]
x = [-1.0, 1.0, 2.0, 1.5, 3.0]
plt.axis('off')
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='off')
plt.bar(x, y)
plt.title('Graph with no border')
plt.show()
```

# Word Cloud
In here, we will use `WordCloud` library at https://github.com/amueller/word_cloud.
World cloud or tag cloud is a common text visualization nowadays. It can show most common or most important words in a website, blog, articles etc.

For example, this is the word cloud of Wikimedia-l generated by IBM word cloud generator.
![Example word cloud](wordcloud_example.png)

Let's make one for city U in Python and Matplotlib. Here, we import the library via `from wordcloud import WordCloud`, read string from file with `open` and finally create the world cloud and plot with `plt.show()`

```python
from wordcloud import WordCloud
text = ""
with open("cityu_english.txt", "r") as f:
    text = f.read()
wordcloud = WordCloud(background_color="white").generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
